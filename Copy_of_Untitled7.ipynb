```python
class data:
  def __init__(self, student_name, age, subject):
    self.student_name=student_name
    self.age=age
    self.subject = subject
  def show(self):
    return self.student_name, self.age, self.subject
```

```python
student = data("Ibrahim", 25, "Python")
student.show()
```

```python
import torch
import torch.nn as nn
x= torch.tensor([1.0,2.0,3.0])
print (x)
print (type[x])
y=torch.rand(3,3)
print (y)
print (y.shape)
```

```python
class net:
  def __init__(self):
    self.layer1=nn.Linear(4,8)
    self.layer2=nn.ReLU()
    self.layer3=nn.Linear(8,1)
  def run(self, x):
    x=self.layer1(x)
    x=self.layer2(x)
    x=self.layer3(x)
    return x
net=net()
```

```python
input=torch.rand(1,4)
output=net.run(input)
print (output)
```
